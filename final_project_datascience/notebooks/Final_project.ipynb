```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
new_title=[]
for i in range(0, len(train_dataset)):
    review = re.sub('[^a-zA-Z]', ' ', train_dataset['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [word for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    new_title.append(review)
```

```python
new_title[10]
```

```python
train_dataset['title']=new_title
```
